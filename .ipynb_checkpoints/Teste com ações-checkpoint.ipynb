```python
import yfinance as yf
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
import tensorflow as tf
import pandas as pd
```

```python
ticker ="PETR4.SA"

#df = yf.download(ticker,period="1y", interval="90m")
df = yf.download(ticker)
df=df["2010":]
df = df.dropna()
base = df.iloc[:, 0].values
```

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        name="Fechamento",
        x=df.index,
        y=base,
        line=dict(color='blue', width=1))
)

fig.add_hline(y=1,line_color="blue",line_width=1, line_dash="dash")
fig.add_hline(y=-1,line_color="blue",line_width=1, line_dash="dash")
fig.update_layout(xaxis_rangeslider_visible=True)
fig.update_layout(paper_bgcolor='rgba(0,0,0,0)', plot_bgcolor='rgba(0,0,0,0)')
```

```python
#base2 = base[(len(base) % periodos):]
```

```python
periodos = 60
previsao_futura = 2 # horizonte

X = base[0:(len(base) - (len(base) % periodos))]
X_batches = X.reshape(-1, periodos, 1)

y = base[1:(len(base) - (len(base) % periodos)) + previsao_futura]
y_batches = y.reshape(-1, periodos, 1)

X_teste = base[-(periodos + previsao_futura):]
X_teste = X_teste[:periodos]
X_teste = X_teste.reshape(-1, periodos, 1)
y_teste = base[-(periodos):]
y_teste = y_teste.reshape(-1, periodos, 1)
```

```python
X_batches.shape, y_batches.shape, X_teste.shape, y_teste.shape
```

```python

tf.random.set_seed(42)

tf.__version__
```

```python
entradas = 1
neuronios_oculta = 100
neuronios_saida = 1
```

```python
class Model(tf.keras.Model):
    def __init__(self,entradas,neuronios_oculta,neuronios_saida):
        super().__init__()
        self.rnn = tf.keras.layers.SimpleRNN(units = neuronios_oculta, activation = 'relu', return_sequences = True)
        self.dense = tf.keras.layers.Dense(units = neuronios_saida)
    
    def call(self, inputs):
        x = self.rnn(inputs)
        x = self.dense(x)

        return x
```

```python
model = Model(1,100,1)
otimizador = tf.keras.optimizers.Adam(learning_rate = 0.001)

for epoca in range(1000):
    with tf.GradientTape() as tape:
        out = model(X_batches)
        erro = tf.keras.losses.mean_squared_error(out, y_batches)
    
    # Uma das vantagens de escrever o modelo como uma instância de `tf.keras.Model`
    # é que podemos recuperar a `var_list` com `model.trainable_variables`
    gradientes = tape.gradient(erro, model.trainable_variables)
    otimizador.apply_gradients(zip(gradientes, model.trainable_variables))

    if epoca % 100 == 0:
        print(epoca + 1, ' erro: ', tf.reduce_mean(erro).numpy())
```

```python
previsoes = model(X_teste)
previsoes
```

```python
y_teste2 = np.ravel(y_teste)
previsoes2 = np.ravel(previsoes.numpy())
```

```python
mae = mean_absolute_error(y_teste2, previsoes2)
mae
```

```python
plt.plot(y_teste2, '*', markersize = 10, label = 'Valor real')
plt.plot(previsoes2, 'o', label = 'Previsões')
plt.legend()
```

```python
from keras.models import Sequential,load_model
from keras.layers import Dense,Dropout,LSTM
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from keras.callbacks import EarlyStopping,ReduceLROnPlateau,ModelCheckpoint
import keras
```

```python
ticker ="PETR3.SA"

#df = yf.download(ticker,period="1y", interval="90m")
base = yf.download(ticker)
base=base["2010":"2022-12"]

base.dropna(inplace=True)
base.reset_index(inplace=True)

#base = pd.read_csv("petr4_treinamento.csv")
#base.dropna(inplace=True)
base_treinamento = base.iloc[:,1:2].values

normalizador = MinMaxScaler(feature_range=(0,1))
base_treinamento_normalizada = normalizador.fit_transform(base_treinamento)
```

```python
previsores = []
preco_real = []
anteriores = 180

for i in range(anteriores,len(base_treinamento)):
    
    previsores.append(base_treinamento_normalizada[i-anteriores:i,0])
    preco_real.append(base_treinamento_normalizada[i,0])
    
previsores,preco_real = np.array(previsores),np.array(preco_real)
previsores = np.reshape(previsores,(previsores.shape[0],previsores.shape[1],1))
```

```python
regressor = Sequential()
regressor.add(LSTM(units=100,return_sequences=True,input_shape=(previsores.shape[1],1)))
regressor.add(Dropout(0.3))

regressor.add(LSTM(units=50,return_sequences=True))
regressor.add(Dropout(0.3))

regressor.add(LSTM(units=50,return_sequences=True))
regressor.add(Dropout(0.3))

regressor.add(LSTM(units=50))
regressor.add(Dropout(0.3))

regressor.add(Dense(units=1,activation="sigmoid"))

regressor.compile(optimizer='adam',loss='mean_squared_error',metrics=['mean_absolute_error'])

es = EarlyStopping(monitor="loss",min_delta=1e-10,patience = 10,verbose=1)    
rlr = ReduceLROnPlateau(monitor="loss",factor=0.2,patience=5,verbose=1)
mcp = ModelCheckpoint(filepath="RedePETR4",monitor="loss",save_best_only=True,verbose=1)

regressor.fit(previsores,preco_real,epochs=70,batch_size=32,callbacks=[es,rlr,mcp])
```

```python
ticker ="PETR3.SA"

#df = yf.download(ticker,period="1y", interval="90m")
base_teste = yf.download(ticker)
base_teste=base_teste["2023-01"]

preco_real_teste = base_teste.iloc[:,1:2].values
base_completa = pd.concat((base["Open"],base_teste["Open"]),axis=0)

entradas = base_completa[len(base_completa)-len(base_teste)-180:].values
entradas = entradas.reshape(-1,1)
entradas = normalizador.transform(entradas)
```

```python
X_teste = []

for i in range(180,112):
    X_teste.append(entradas[i-180:i,0])
    

X_teste = np.array(X_teste)
X_teste = np.reshape(X_teste,(X_teste.shape[0],X_teste.shape[1],1))

previsoes = regressor.predict(X_teste)
previsoes = normalizador.inverse_transform(previsoes)
```

```python
preco_real_teste.mean()
```

```python
plt.plot(preco_real_teste,color='red',label="Preço real")
plt.plot(previsoes,color='blue',label="Previsoes")

plt.title("Previsão do preço das ações")
plt.xlabel("Tempo")
plt.ylabel("Valor Yahoo")
plt.legend()
plt.show()
```

```python
df_results = pd.DataFrame()

df_results["Preço real"] = preco_real_teste.reshape(-1)
df_results["Previsao"] = previsoes.reshape(-1)
df_results["Diferenca"] = df_results["Preço real"]-df_results["Previsao"]
```
